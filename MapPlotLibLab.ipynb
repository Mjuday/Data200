```python
import matplotlib.pyplot as plt
```

```python
Hours_studied = [1,2,4,5,6,7,8,9,10,11]
ExamScores = [53, 59, 66, 73, 76, 79, 81, 90, 93, 97]

plt.scatter(Hours_studied, ExamScores, label = 'students vs test scores', color = 'red', marker = '*')
plt.plot(Hours_studied, ExamScores, color = 'k')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Scores')
plt.title('Sudent Studying verses Exam Scores')
plt.legend(loc="best")
plt.grid()
plt.show()
```

```python

data = [600, 150, 300, 200, 150]
data_labels = ['Groceries $', 'Entertainment $', 'Utilities $', 'Transportation $', 'Health $']

plt.figure(figsize = (5,5))
plt.pie(data, labels = data_labels, colors = ['red', 'brown', 'blue', 'yellow', 'orange'], autopct = '%.2f%%')
plt.title('Spending for the Month')
plt.show()
```
