```python
x = np.linspace(-2*np.pi, 2*np.pi,100)
x
```

```python
sinx = np.sin(x)
sinx
```

```python
cosx = np.cos(x)
x
```

```python
plt.plot(x,sinx, color = 'blue', linestyle = '--', label = 'sin', x,cosx, color = 'red', linestyle = '-', label = 'cos')
plt.show()
```

```python
plt.plot(x,sinx,label = 'sin', color = 'blue', linestyle = '--')
plt.plot(x,cosx, label = 'cos', color = 'red', linestyle = '-')
plt.xlabel('Radian')
plt.ylabel('Amplitude')
plt.title('My First Plot')
plt.legend(loc = 'best')
plt.grid()
plt.show()
```

```python
plt.plot(x,sinx, '*-.r', label = 'sin',markeredgewidth = 1, markeredgecolor = 'k', markersize = 10)
plt.xlabel('Radian')
plt.ylabel('Amplitude')
plt.title('My First Plot')
plt.legend(loc = 'best')
plt.grid()
plt.show()
```

```python
# plot sin and cos in two different subplots but in one window(figure)
#subplot 1: sin
plt.subplot(2,1,1) #2 is number of rows in the plot, 1 is the number of columns, 1 is the position
plt.plot(x,sinx,'*--r', label = 'sin')
plt.grid()
plt.legend()
plt.xlabel('Radian')
plt.ylabel('Amplitude')

#subplot 2: cos
plt.subplot(2,1,2)
plt.plot(x,sinx,'o--g', label = 'cos')
plt.grid()
plt.legend()
plt.xlabel('Radian')
plt.ylabel('Amplitude')

plt.show()
```

### plot types

- line chart
- semilog plot
- bar chart
- histogram
- scatterplot


```python
x = np.linspace(0.01, 5, 100)
x
```

```python
y = np.exp(x)
y
```

```python
#linear plot
plt.plot(x,y)
plt.xlabel(x)
plt.ylabel(y)
plt.title("Linear space")
plt.show()
```

```python
# semilog
plt.semilogy(x,y)
plt.xlabel(x)
plt.ylabel(y)
plt.title("Logarithmic space")
plt.show()
```
