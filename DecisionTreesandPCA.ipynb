```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import tree
from matplotlib import pyplot as plt
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

# Decision Trees

## with scikit learn using Iris dataset

```python
iris = load_iris()
x = iris.data
y = iris.target
x_train, x_test, y_train, y_test = train_test_split(x,y)
clf = DecisionTreeClassifier(max_leaf_nodes = 3, criterion = "entropy")
clf.fit(x_train, y_train)
```

```python
tree.plot_tree(clf)
plt.show()
```

```python
clf.score(x_train, y_train)
```

## with sklearn using digits dataset

```python
digits = load_digits()
x1 = digits.data
y1 = digits.target
x_train1, x_test1, y_train1, y_test1 = train_test_split(x1,y1)
clf1 = DecisionTreeClassifier(criterion = "entropy")
clf1.fit(x_train1, y_train1)
```

```python
tree.plot_tree(clf1)
plt.show()
```

```python
clf1.score(x_train1, y_train1)
```

# PCA

## dimensional reduction for iris dataset from 4D to 2D

```python
x2 = iris.data
y2 = iris.target
target_names = iris.target_names
```

```python
# tsandardize the data
scaler = StandardScaler()
x2_scaler = scaler.fit_transform(x2)
```

```python
# Principle Component Analysis (PCA) transform the data from 4D to 2D
pca = PCA(n_components = 2)
x2_pca = pca.fit_transform(x2_scaler)
```
