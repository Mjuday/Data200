```python
plt.plot(x,np.sin(x),label='sin',color='blue',linestyle='--')
plt.plot(x,np.cos(x),label='cos',color='red',linestyle='-')
plt.xlabel('x')
plt.ylabel('y')
plt.title('My First Plot')
plt.legend(loc="best")
plt.grid()
plt.show()
```

```python
x = np.linspace(-2*np.pi,2*np.pi,100)
sinx = np.sin(x)
plt.plot(x,sinx,'*--r',markersize=10,markeredgewidth=1,markeredgecolor='k',label='sin')
plt.legend(loc="best")
plt.xlabel("Radian")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```

```python
#Mltiple plots in the same window
x = np.linspace(-2*np.pi,2*np.pi,100)
sinx = np.sin(x)
cosx = np.cos(x)
####### subplot ########
plt.subplot(2,1,1) #divide figure to two rows and on column, use location 1 on top
plt.plot(x,sinx,'*--r',label='sin')
plt.grid()
plt.legend()
plt.xlabel("Radian")
plt.ylabel("Amplitude")

plt.subplot(2,1,2) #divide figure to two rows and on column, use location 2 on down
plt.plot(x,cosx,'o-g',label='cos')
plt.grid()
plt.legend()
plt.xlabel("Radian")
plt.ylabel("Amplitude")

plt.show()
```

# Plot Types

1. Semilog Plot

```python
x = np.linspace(0.01,5,100)
x
```

```python
e = np.exp(x)
e
```

```python
#Linear plot
plt.plot(e,x)
plt.xlabel("x")
plt.ylabel("y=exp(x)")
plt.title("Linear Y axis")
```

```python
plt.semilogy(x,e)
plt.xlabel("x")
plt.ylabel("y=exp(x)")
plt.title("Log Y axis")
```

2. Histogram

```python
ur = np.random.rand(10000)
nr = np.random.randn(10000)

plt.subplot(2,2,1)
plt.hist(ur)
plt.xlabel("Uniform Random Number, Default 10 Bin")

plt.subplot(2,2,2)
plt.hist(ur,20)
plt.xlabel("Uniform Random Number, 20 Bin")

plt.subplot(2,2,3)
plt.hist(nr)
plt.xlabel("Normal Random Number Default 10 Bin")

plt.subplot(2,2,4)
plt.hist(nr,20)
plt.xlabel("Normal Random Number, 20 Bin")
```

3. Scatter Plot

```python
#Scatter plot
x = np.linspace(0,2*np.pi,100)
#noise = np.random.randn(100) #Next
signal = 2*np.sin(x)
#y = signal + 0.2* noise #Next
plt.plot(x,signal)
#plt.scatter(x,y)
plt.show()
```

```python
x = np.linspace(0,2*np.pi,100)
noise = np.random.randn(100) #Next
signal = 2*np.sin(x)
y = signal + 0.2* noise #Next
plt.plot(x,signal)
plt.scatter(x,y)
plt.show()
```

4. Pie Chart

```python
x = [30,20,15,25,10]
data_name = ['data1','data2','data3','data4','data5']
plt.figure(figsize=(5,5))
plt.pie(x,labels=data_name)
plt.show()
```

```python
#Pie Plot
x = [30,20,15,25,10]
data_name = ['data1','data2','data3','data4','data5']
plt.figure(figsize=(5,5))
plt.pie(x,explode = [0.01,0,0,0.04,0.09],labels=data_name,autopct='%.2f%%')
plt.show()
```

5. Polar Chart

```python
x = np.linspace(0,2*np.pi,1000)
y = np.cos(2*x)
plt.polar(x,y)
plt.show()
```

6. Bar Chart

```python
#Bar Chart
years = [1999,2014,2030]
x=[1,2,4]
plt.bar(years,x)
plt.show()
```

```python
x = [1,2,4]
y = [3.5,3,2]
z = [2,3,1.5]

width = 0.2
locs= np.arange(1,len(x)+1)
plt.bar(locs,x,width=width,label='x')
plt.bar(locs+width,y,width=width,color='red',label='y')
plt.bar(locs+2*width,z,width=width,color='black',label='z')
plt.legend()
plt.xticks([1.25,2.25,3.25],['2000','2005','2010'])
plt.show()
```
